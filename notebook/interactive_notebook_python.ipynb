```python
# library
import pandas as pd
import numpy as np
import pickle
from sklearn.datasets import load_breast_cancer

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.svm import SVC, LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load the breast cancer dataset from sklearn's datasets module
data = load_breast_cancer()

# create dataframe
df = pd.DataFrame(data.data, columns=data.feature_names)

# target labels (1 for benign/jinak, 0 for malignant/ganas)
df['class'] = data.target

# display dataframe
df.head()
```

```python
# information
df.info()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicated data
print(df.duplicated().sum())
```

```python
# df['class'] = df['class'].apply(lambda x: -1 if x == 0 else x)
```

```python
# target labels (1 for benign/jinak, 0 for malignant/ganas)
df['class'].value_counts()
```

```python
# split
X = df.drop('class', axis=1)
y = df['class']

X = StandardScaler().fit_transform(X)
X = pd.DataFrame(X, columns=data.feature_names)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# object model
svm_model = SVC()
svml_model = LinearSVC()
rf_model = RandomForestClassifier(random_state=42)
mlp_model = MLPClassifier(max_iter=1000)

svm_model.fit(X_train, y_train)
svml_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
mlp_model.fit(X_train, y_train)

svm_predictions = svm_model.predict(X_test)
svml_predictions = svml_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)
mlp_predictions = mlp_model.predict(X_test)

svm_accuracy = accuracy_score(y_test, svm_predictions)
svml_accuracy = accuracy_score(y_test, svml_predictions)
rf_accuracy = accuracy_score(y_test, rf_predictions)
mlp_accuracy = accuracy_score(y_test, mlp_predictions)

print(f"SVM Accuracy: {svm_accuracy}")
print(f"Linear SVM Accuracy: {svml_accuracy}")
print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"MLP Accuracy: {mlp_accuracy}")
```

```python
# report
print("SVM Confusion Matrix:")
print(confusion_matrix(y_test, svm_predictions))
print("\nSVM Classification Report:")
print(classification_report(y_test, svm_predictions))

print("\nLinear SVM Confusion Matrix:")
print(confusion_matrix(y_test, svml_predictions))
print("\nLinear SVM Classification Report:")
print(classification_report(y_test, svml_predictions))

print("\nRandom Forest Confusion Matrix:")
print(confusion_matrix(y_test, rf_predictions))
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, rf_predictions))

print("\nMLP Confusion Matrix:")
print(confusion_matrix(y_test, mlp_predictions))
print("\nMLP Classification Report:")
print(classification_report(y_test, mlp_predictions))
```

```python
def validate_model(model, parameter_set, train_data, test_data):
    clf = GridSearchCV(estimator=model, param_grid=parameter_set, n_jobs=3, cv=10)

    clf.fit(train_data[0], train_data[1])

    grid_scores = clf.cv_results_
    best_score = clf.best_score_
    best_params = clf.best_params_
    best_estimator = clf.best_estimator_
    test_score = clf.score(test_data[0], test_data[1])

    return grid_scores, best_score, best_params, best_estimator, test_score
```

```python
svm_parameter_set = {
    'C': [0.01, 0.1, 1, 10, 100, 1000],
    'gamma': ['scale', 'auto', 1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
    }

svml_parameter_set = {
    "loss": ("hinge", "squared_hinge"),
    "C": [0.1, 1, 10, 100, 1000, 10000, 5, 50, 500, 5000]
    }

rf_parameter_set = {
    "n_estimators": [50, 100, 200, 500],
    "max_depth": [None, 10, 20, 30, 50],
    "max_features": ["auto", "sqrt", "log2"],
    "criterion": ["gini", "entropy", "log_loss"]
}

mlp_parameter_set = {
    "activation": ["identity", "logistic", "tanh", "relu"],
    "solver": ["sgd", "adam"],
    "batch_size": [16, 32, 64, 128],
    "hidden_layer_sizes": [(16,), (32,), (64,), (128,)],
    "learning_rate": ["constant", "invscaling", "adaptive"],
    "learning_rate_init": [1e-1, 1e-2, 1e-3],
    "max_iter": [200, 300, 500, 1000, 2000]
    }
```

```python
svm_grid_model = SVC()
grid_scores, best_score, best_params, best_estimator, test_score = validate_model(
    model=svm_model,
    parameter_set=svm_parameter_set,
    train_data=[X_train, y_train],
    test_data=[X_test, y_test],
)
```

```python
print(f"SVM grid search cv: {grid_scores}")
print(f"SVM best score: {best_score}")
print(f"SVM best params : {best_params}")
print(f"SVM best estimator : {best_estimator}")
print(f"SVM test score : {test_score}")
```
